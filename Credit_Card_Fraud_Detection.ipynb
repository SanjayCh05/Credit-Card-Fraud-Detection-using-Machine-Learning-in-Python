```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
#LOADING THE DATA TO PANDA DATAFRAME

credit_card_data= pd.read_csv('/content/creditcard.csv')
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# checking the number of missing values in each row
credit_card_data.isnull().any(axis=1)
```

```python
#fixing the missing values by removing the row
credit_card_data.drop(labels=np.where(credit_card_data.isnull().any(axis=1))[0],axis=0,inplace=True)
```

```python
credit_card_data.isnull().any(axis=1)
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

This Dataset is highly unblanced

0 --> Normal Transaction

1 --> fraudulent transaction

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
X = credit_card_data.drop('Class', axis =1)
y = credit_card_data['Class']
```

Splitting the data into training data and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,  stratify=y, random_state=2)
```

Balancing the dataset using SMOTE

```python
smote = SMOTE(sampling_strategy=0.5, random_state=42)
X_smote, y_smote = smote.fit_resample(X_train, y_train)
print(y_train.value_counts())
print(y_smote.value_counts())
```

```python
# Scaling the Training Dataset
scaler = StandardScaler()
X_smote_scaled = scaler.fit_transform(X_smote)
X_test_scaled = scaler.transform(X_test)
```

LOGISTIC REGRESSION

```python
model=LogisticRegression(solver='newton-cg',max_iter=8000)
```

```python
model.fit(X_smote_scaled, y_smote)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_smote_scaled)
training_data_accuracy = accuracy_score(X_train_prediction, y_smote)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test_scaled)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
```
